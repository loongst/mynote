替换kval目录下kval值为空的项

```python
import re 
import os
regx=re.compile(r"\s{6}-\stype:\skval\n\s{8}kval:\n")
# regxfind=re.compile(r"kval:\s{0,6}\n\s{6}-")
for i in os.listdir("kval"):
    with open("kval/{}".format(i),"+r",encoding="utf-8") as f1,open("kval/{}.bak".format(i),"+w",encoding="utf-8") as f2:
        con=f1.read()
        new=re.sub(regx,"",con)
        print(new)
        f2.write(new)
    os.remove("kval/{}".format(i))
    os.rename("kval/{}.bak".format(i),"kval/{}".format(i))
```
